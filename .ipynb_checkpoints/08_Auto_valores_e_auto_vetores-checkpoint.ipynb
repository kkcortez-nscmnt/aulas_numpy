# Auto Valor e auto vetor

|Função(A)|Descrição|
|:-|:-|
|linalg.eig(A)| retorna os autovalores e autovetores de uma matrix A|

### No sympy


```python
import numpy as np
```

```python
A = np.array([[5,3,-1],
             [3,4,0],
             [-1,0,2]])
A
```

```python
l,v = np.linalg.eig(A)
```
