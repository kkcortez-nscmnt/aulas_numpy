# Métodos estatísticos para arrays

|Método|Descrição|
|:-|:-|
|.max() / .min()|retornam os valores máximo e mínimo|
|.argmax() / .argmin()|retorna os índices de valor máximo e mínimo
|.sum() / .mean() / .std()| soma, média e desvio padrão
|.cumsum()|retorna um array com a soma acumulada|



```python
import numpy as np

my_array = np.arange(8).reshape(2,4)**2
my_array[0][0] = 11
my_array
```

```python
print(my_array.max())
print(my_array.argmax())
print(my_array.argmin())
print(my_array.sum())
print(my_array.mean())
print(my_array.std())
print(my_array.cumsum())
```
