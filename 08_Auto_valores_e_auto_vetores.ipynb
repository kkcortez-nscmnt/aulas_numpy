# linalg

|Função(A)|Descrição|
|:-|:-|
|linalg.eig(A)| retorna os autovalores e autovetores de uma matrix A|
|linalg.solve(A,B)| resolve um sistema  do tipo AX=B|

### Auto Valores e Auto Vetores


```python
import numpy as np
```

```python
A = np.array([[5,3,-1],
             [3,4,0],
             [-1,0,2]])
A
```

```python
l,v = np.linalg.eig(A)
```

```python
l
```

```python
v
```

# Solver para sistemas lineares

```python
B = np.array([[1,1,1],[1,2,2],[2,1,3]])
B
```

```python
C= np.array([6,9,11])
C
```

```python
np.linalg.solve(B,C)
```
