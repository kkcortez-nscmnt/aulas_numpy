# Método copy

Ao usarmos um sinal de igual não copiamos um array. O array é um objeto e possui um método para ser copiado

|Método|Descrição|
|:---:|:---:|
|.copy()| faz a cópia(deep copy) do array. OBS!! O sinal de igual não executa essa função|

Nota - a propiedade __*base*__ retorna o array utilizado para criar o array em questão

```python
import numpy as np
a = np.arange(8)
```

```python
b = a
b
```

```python
a[0] = 9
a
```

```python
b 
```

```python
a is b
```

```python
b == a
```

## Para copiarmos um array utilizamos o método copy

```python
c = a.copy()
c 
```

```python
a[0] = 1
print(a)
print(b)
print(c)
```

### Arrays podem compatilhar as mesmas propiedades, mesmo sendo arrays diferentes

```python
d = a.reshape(2,4)
d
```

```python
print(d is a)
print(d.base is a)
print(a.base)
```
