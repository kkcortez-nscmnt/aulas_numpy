# Índices e fatias de arrays

## indexing e slicing

|Sintaxe|Descrição|
|:-|:-|
|[i]| para 1 dimensão funciona como nas listas|
|[:i]| exibe um array que vai do indice 0 até o anterior ao i|
|[i:]| exibe um array que vai do indice posterior ao i até o último indice|
|[::i]| exibe todos os elementos variando de i em i|
|[::-i]| exibea sequência de elementos de trás pra frente variando de i em i|
|[a:b:c]|início, fim, step|

Para arrays multidimensionais:

|Sintaxe|Descrição|
|:-|:-|
|[i:j]| para exibir um elemento de uma matriz|
|[i,a:b]| retorna uma fatia da linha i com elementos das colunas de a até b
|[n,i,j]| outra dimensão, linha, coluna


```python
import numpy as np

a = np.arange(10)**2
a
a[-1]
```

```python
a[1]
```

```python
a[:2]
```

```python
a[2:]
```

```python
a[::2]
```

```python
a[2::]
```

```python
a[::-1]
```

```python
a[2:8:2]
```

## Para Arrays multidimensionais

```python
b = np.arange(20).reshape(4,5)
b
```

Podemos criar um "sub array" de b

```python
b[0]
```

```python
b[1,1]
```

```python
b[1,2:]
```

```python
b[2:,3:]
```

```python
c  = np.arange(40).reshape(2,4,5)
c
```

```python
c[1,1,1]
```
