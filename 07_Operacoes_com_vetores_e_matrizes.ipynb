<h1> Operações com vetores e matrizes no Numpy

|Funções|Descrição|
|:-|:-|
|dot(a,b) ou @| retorna o produto escalar entre os vetores a e b|
|cross(a,b)| retorna o produto vetorial entre os vetores a e b|
|inner(a,b) e outer(a,b)|retorna o produto interno e externo entre os vetores a e b
|linag.norma(a)| retorna a norma do vetor a|
|matmul(a,b) ou @| produto entre as matrizes a e b
|linalg.det(A)| retorna o determinante de matriz A
|linalg.inv(A)| retorna a matriz inversa  de A


# Vetores

```python
import numpy as np
import sympy as sp
```

```python
a = np.array([1, 2, 3])
print(a.shape)
print(a.ndim)
```

# Produto escalar entre vetores

```python
np.dot(a,a)
```

a_s = sp.Matrix([1,2,3])
print(a_s.T*a_s) # a transporta vezes o vetor retorna um escalar
a_s*a_s.T # o vetor vezes a transposta retorna uma matrix nxn



# Produto vetorial entre dois vetores

O produto vetorial entre dois vetores retorna, em termos geométricos, o vetor ortogonal a esses dois vetores

```python
np.cross(a,a)
```

```python
v_1 =np.array([1,0,0])
v_2 =np.array([0,1,0])
np.cross(v_1,v_2)
```

# Inner e outer

```python
np.inner(a,a)
```

```python
np.outer(a,a)
```

### Utilizando o Numpy

```python
A = np.array([[1,2],[4,4]])
A
```

```python
np.matmul(A,A)
```

```python
A@A
```
